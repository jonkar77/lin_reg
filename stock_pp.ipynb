```python
#install the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#load the data
df = pd.read_csv('/content/drive/MyDrive/machine learning/dataset/b_dataset.csv')
```

```python
#define the features and target
X = df[['Open','Volume']]
y = df['Close']
```

```python
#test-train-split-20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
```

```python
#fit the model in the linear regression
regressor.fit(X_train, y_train)
```

```python
#predict the test values
y_pred = regressor.predict(X_test)
```

```python
#evaluate the performance
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# plot the actual vs predicted values
# plot is useful for evaluating how well the model performs on new data
actual_colors = np.where(y_test > y_pred, 'r', 'g')
plt.scatter(y_test, y_pred, c=actual_colors)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Prices')
plt.show()
```

```python
# plot the residuals
# this plot helps to check if the assumptions of linear regression are met.
actual_colors = np.where(y_test > y_pred, 'r', 'g')
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, c=actual_colors)
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
